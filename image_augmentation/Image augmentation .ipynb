```python
import numpy as np
import matplotlib.pyplot as plt
import skimage as sk
import scipy
```

```python
img = plt.imread("Cat.webp")
plt.imshow(img)
```

```python
img.shape
```

```python
#1 Rotation of the image
degree = 30
imx = sk.transform.rotate(img,degree)
plt.imshow(imx)
```

```python
#2 Rescale the image
scale = (0.3,0.3)
imx = sk.transform.rescale(img, scale)
plt.imshow(imx)
```

```python
#3 Color invert
imx = sk.util.invert(img)
plt.imshow(imx)
```

```python
#4 Cropping to the face
hl,hu = 100,240                      # Horizontal lower and upper value
vl,vu = 160,320                      # Verticle lower and upper value
imx = img[hl:hu,vl:vu]
plt.imshow(imx)
```

```python
#5 Flipping up down
imx = np.flipud(img)
plt.imshow(imx)
```

```python
#6 Flipping left right
imx = np.fliplr(img)
plt.imshow(imx)
```

```python
#7 Flip updown as well as right left
imx = np.flip(img)
plt.imshow(imx)
```

```python
#8 Adding random noise

imx = sk.util.random_noise(img)
plt.imshow(imx)
```

```python
#9 Changing the intesity, contrast
minint = 10
maxint = 120
imx = sk.exposure.rescale_intensity(img, in_range=(minint,maxint))
plt.imshow(imx)
```

```python
#10 Blurred image
filtersize = (11,11,1)
imx = scipy.ndimage.uniform_filter(img,filtersize)
plt.imshow(imx)
```

```python
#11 RGB to XYZ color space conversion
imx = sk.color.rgb2xyz(img)
plt.imshow(imx)
```

```python
#12 Gamma correction on image
gamma = 0.4
imx = sk.exposure.adjust_gamma(img, gamma)
plt.imshow(imx)
```

```python
#13 Sigmoid correction on image 
imx = sk.exposure.adjust_sigmoid(img)
plt.imshow(imx)
```

```python
#14 Sigmoid correction on image with color inversion
imx = sk.exposure.adjust_sigmoid(img, inv=True)
plt.imshow(imx)
```

```python
#15 Rotate and flip
degree = 40
imx = sk.transform.rotate(img,degree)
imx = np.flip(imx)
plt.imshow(imx)
```

```python
#16 Blur and flip
filtersize = (11,11,1)
imx = scipy.ndimage.uniform_filter(img,filtersize)
imx = np.flip(imx)
plt.imshow(imx)
```

```python
#17 Blur and rotate
filtersize = (11,11,1)
rotation = 30
imx = scipy.ndimage.uniform_filter(img,filtersize)
imx = sk.transform.rotate(imx,rotation)
plt.imshow(imx)
```

```python
#18 crop and flip
hl,hu = 100,240                      # Horizontal lower and upper value
vl,vu = 160,320                      # Verticle lower and upper value
imx = img[hl:hu,vl:vu]
imx = np.flip(imx)
plt.imshow(imx)
```

```python
#19 noise and flip

imx = sk.util.random_noise(img)
imx = np.fliplr(imx)
plt.imshow(imx)
```

```python
#20 Changing intensity and rotate and flip
minint = 10
maxint = 130
degree = 30
imx = sk.exposure.rescale_intensity(img, in_range=(minint,maxint))
imx = sk.transform.rotate(imx,degree)
imx = np.flipud(imx)
plt.imshow(imx)
```
